#### Step-2(Importing Necessary Modules)

```python
import cv2 as cv
import os
```

#### Step-3(Defining and loading haar cascade path)

```python
har_path = cv.data.haarcascades + "haarcascade_frontalface_default.xml"
```

```python
#loading the haar cascade classifier
detects = cv.CascadeClassifier(har_path)
#Checking the cascade is empty or not
if detects.empty():
    raise RuntimeError(f"Failed to load cascade from: {har_path}")
else:
    print("CasCade Classifier Successfully Loaded")
```

#### Step-4(writing code to open webcam and detect face)

```python
#defining the function to open webcam untill press q and detecting the face
def detect_f():
    #Opening the camera using the vediocapture()
    webcam = cv.VideoCapture(0)
    #Checking the Web Cam is Opened or not
    if webcam.isOpened():
        print("WebCam is Opened Successfully and Press 'q' to Exit")
    else:
        print("Error:WebCam is not Opened check Camera Permissions")
    try:
        while True:
            #Reading the web Cam 
            ret,frame = webcam.read()
            if not ret:
                print("Error: Frame Captured Failed")
                break
            #Converting to the GreyScale Images
            greys = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
            #Detecting the Faces on the Web Cam
            faces = detects.detectMultiScale(greys,scaleFactor=1.1,minNeighbors=5)
            #Drawing the Rectangles on the Faces
            for x,y,w,h in faces:
                # Draw rectangles using the faces as (x,y) top-left, (x+w,y+h) bottom-right
                cv.rectangle(frame,(x,y),(x+w,y+h),(0,250,0),3)
            # Display face count
            cv.putText(frame,f"Faces:{len(faces)}",(10, 25),cv.FONT_HERSHEY_SIMPLEX,0.7,(0,255,0),2)
            #Displaying the frame using imshow()
            cv.imshow("SSA Web Cam",frame)
            #Exit on pressing the q
            if cv.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        #Cleaning the Resource
        webcam.release()
        cv.destroyAllWindows()
        print("Web Cam Released and Windows Closed")
```

```python
#Calling the Web Cam Function
detect_f()
```

#### Step-5(Saving the Captured Images)

```python
#Defining the parameters for the saving the captured images
save_image = True
Images_D = "SSA Images"
#Creating the directory and saving the images 
if save_image:
    os.makedirs(Images_D,exist_ok=True)
```

```python
#Defining the function to play a short beep when an image is saved
def play_beep():
    try:
        #This is the Windows built-in module
        import winsound
        #Defining the frequency as 1000Hz and duration 150ms
        winsound.Beep(1000,150)
    except Exception:
        #fallback no-op that is we avoid adding new heavy deps
        pass
```

```python
#Importing the time and date modules to saving the image with time and date
import time
from datetime import datetime
#defining the function to open webcam untill press q and detecting the face
def detect_faces():
    #Opening the camera using the vediocapture()
    webcam = cv.VideoCapture(0)
    #Checking the Web Cam is Opened or not
    if webcam.isOpened():
        print("WebCam is Opened Successfully and Press 'q' to Exit")
    else:
        print("Error:WebCam is not Opened check Camera Permissions")
    try:
        while True:
            #Reading the web Cam 
            ret,frame = webcam.read()
            if not ret:
                print("Error: Frame Captured Failed")
                break
            #Converting to the GreyScale Images
            greys = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
            #Detecting the Faces on the Web Cam
            faces = detects.detectMultiScale(greys,scaleFactor=1.1,minNeighbors=5)
            #Drawing the Rectangles on the Faces
            for x,y,w,h in faces:
                # Draw rectangles using the faces as (x,y) top-left, (x+w,y+h) bottom-right
                cv.rectangle(frame,(x,y),(x+w,y+h),(0,250,0),3)
            # Display face count
            cv.putText(frame, f"Faces: {len(faces)}", (10, 25), cv.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0), 2)
            # Save cropped intruder(s) using simple cooldown to avoid duplicates
            now = time.time()
            if save_image and len(faces) > 0 and (now - 0) > 15:
                last_save_time = now
                ts = datetime.now().strftime("%Y%m%d_%H%M%S")
                # Save all detected face crops (or change to save only largest)
                for idx, (x, y, w, h) in enumerate(faces, start=1):
                    unique_id = str(int(time.time() * 1000))  # Unique based on timestamp
                    fname = os.path.join(Images_D, f"intruder_{ts}_{unique_id}_{idx}.jpg")
                    cv.imwrite(fname,frame)
                print(f"[{datetime.now().strftime('%H:%M:%S')}] Saved {len(faces)} images(s) to '{Images_D}'")
                play_beep()
            #Displaying the frame using imshow()
            cv.imshow("SSA Web Cam",frame)
            #Exit on pressing the q
            if cv.waitKey(1) & 0xFF == ord('q'):
                break
    finally:
        #Cleaning the Resource
        webcam.release()
        cv.destroyAllWindows()
        print("Web Cam Released and Windows Closed")
```

```python
detect_faces()
```
